## Circuit Fundamentals

BJT circuits require confident use of Ohm's law, KVL, KCL, voltage dividers, and power calculations.

---

**Q1.** You have a 9V supply, a 1 kΩ resistor, and a component that drops 0.7V, all in series. Find: (a) the current, (b) the voltage across the resistor, (c) the power dissipated in the resistor.

<details>
<summary>Answer</summary>

(a) I = (9V − 0.7V) / 1kΩ = 8.3 mA

(b) V_R = 9V − 0.7V = 8.3V (or V_R = IR = 8.3 mA × 1kΩ = 8.3V)

(c) P = I²R = (0.0083)² × 1000 = 68.9 mW

This is a common BJT base-drive calculation: the 0.7V component is V_BE.

*Reference: Module 00, Notebook 01 — Ohm's law, KVL, Power sections*
</details>

---

**Q2.** (KCL) At a circuit node, 10 mA flows in from wire A and 3 mA flows in from wire B. How much current flows out through wire C (the only other connection)?

<details>
<summary>Answer</summary>

KCL: current in = current out. 10 mA + 3 mA = 13 mA flows out through wire C.

In a BJT: I_E = I_C + I_B. The emitter node is exactly this kind of KCL node.

*Reference: Module 00, Notebook 01 — KCL section*
</details>

---

**Q3.** A voltage divider has R1 = 10 kΩ (top) and R2 = 2.2 kΩ (bottom), connected to a 12V supply. What is the voltage at the middle node (across R2)? If you connect a load to this node, what happens to the voltage and why?

<details>
<summary>Answer</summary>

V_out = 12V × R2/(R1+R2) = 12 × 2200/12200 = 2.16V.

Connecting a load puts a resistance in parallel with R2, reducing the effective bottom resistance. This pulls V_out lower than the unloaded value. The heavier the load (lower resistance), the more V_out drops. This is why voltage dividers work best with high-impedance loads.

In BJT circuits, voltage dividers set the base bias voltage. The base draws current, which loads the divider — this is a key design consideration.

*Reference: Module 00, Notebook 01 — Voltage divider section*
</details>

## Power Dissipation

Transistors dissipate power. Exceeding the maximum power rating destroys them. You need to be comfortable calculating power.

---

**Q4.** A component has 6V across it and 50 mA flowing through it. How much power does it dissipate? If its maximum rating is 625 mW, is it within spec?

<details>
<summary>Answer</summary>

P = V × I = 6V × 50 mA = 300 mW. Yes, 300 mW < 625 mW — it's within spec.

The 625 mW maximum is the 2N3904's power rating. You'll use this check whenever designing BJT circuits.

*Reference: Module 00, Notebook 01 — Power section*
</details>

---

**Q5.** Why does power grow as I² (not just I) when current through a fixed resistance doubles?

<details>
<summary>Answer</summary>

P = I²R. Doubling I means the voltage across R also doubles (V = IR), so both V and I double: P = V × I → (2V)(2I) = 4× the original power. Power grows with the **square** of current. This is why exceeding rated current is so dangerous — a 2× overcurrent causes 4× the heat.

*Reference: Module 00, Notebook 01 — Power formula intuition section*
</details>

## PN Junctions and Diodes

A BJT contains **two** PN junctions. You must understand how a single PN junction works before you can understand a transistor.

---

**Q6.** What is the **depletion region** in a PN junction? Why does it form, and what stops it from growing indefinitely?

<details>
<summary>Answer</summary>

When P-type meets N-type silicon, electrons diffuse from the N-side into the P-side and recombine with holes (and vice versa). This leaves behind fixed ions: positive donor ions on the N-side, negative acceptor ions on the P-side. These ions create an electric field that opposes further diffusion. Equilibrium is reached when the electric field exactly balances the diffusion tendency — that's what stops it from growing.

*Reference: Module 02, Notebook 01 — "What Happens When P Meets N" section*
</details>

---

**Q7.** What is the built-in potential of a silicon PN junction? What happens to the depletion region when you apply forward bias? Reverse bias?

<details>
<summary>Answer</summary>

The built-in potential is approximately **0.6–0.7V** for silicon. Forward bias (positive on P-side) opposes the built-in potential, shrinks the depletion region, and allows current to flow. Reverse bias (positive on N-side) adds to the built-in potential, widens the depletion region, and blocks current (except for tiny leakage).

*Reference: Module 02, Notebook 01 — "Forward and Reverse Bias" section*
</details>

---

**Q8.** A silicon diode is in series with a 220Ω resistor, powered by 5V. Assuming V_F = 0.7V, what current flows? What if the diode is reversed — how much current flows then?

<details>
<summary>Answer</summary>

Forward: I = (5V − 0.7V) / 220Ω = 19.5 mA.

Reversed: essentially zero (a few nanoamps of leakage). The reverse-biased diode acts as an open circuit for practical purposes.

*Reference: Module 02, Notebook 02 — Diodes in Practice*
</details>

## Doping and Carriers

The BJT's operation depends on **minority carrier injection** across a thin base region. You need to understand what majority and minority carriers are.

---

**Q9.** What are the majority carriers in N-type silicon? In P-type silicon? What are the minority carriers in each?

<details>
<summary>Answer</summary>

N-type: majority = electrons, minority = holes. P-type: majority = holes, minority = electrons. Minority carriers exist due to thermal generation and are far less abundant than majority carriers.

*Reference: Module 02, Notebook 01 — N-type and P-type doping sections*
</details>

---

**Q10.** Phosphorus has 5 valence electrons. When it replaces a silicon atom in the crystal, what happens to the 5th electron? Why is the silicon still electrically neutral despite having this "extra" electron?

<details>
<summary>Answer</summary>

The 5th electron has no bond to join (silicon only forms 4 bonds), so it is loosely bound and easily freed into the conduction band. The silicon is still neutral because the phosphorus atom brought 15 protons along with its 15 electrons — no net charge was added, just a mobile carrier.

*Reference: Module 02, Notebook 01 — "N-Type Doping" section*
</details>

## Putting It Together

---

**Q11.** Consider this circuit: a 9V supply, a 470Ω resistor (R_C), and a PN junction in series. The junction drops 0.7V when forward-biased. Calculate I, V across R_C, and power dissipated in (a) the resistor and (b) the junction.

<details>
<summary>Answer</summary>

I = (9V − 0.7V) / 470Ω = 17.7 mA

V_RC = 9V − 0.7V = 8.3V

P_resistor = I × V_RC = 17.7 mA × 8.3V = 146.7 mW

P_junction = I × V_junction = 17.7 mA × 0.7V = 12.4 mW

Total power: 146.7 + 12.4 = 159.1 mW = V_supply × I = 9V × 17.7 mA ✓

This is the pattern for nearly every BJT calculation: replace "PN junction" with "base-emitter junction," add a collector resistor, and you have a transistor circuit.
</details>
