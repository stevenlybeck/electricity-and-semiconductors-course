## Ohm's Law and Circuit Analysis

Module 02 involves calculating currents through diodes and resistors. You need Ohm's law to be second nature.

---

**Q1.** A 330Ω resistor is connected in series with a component that drops 0.7V, powered by a 5V source. What is the current in the circuit? What is the voltage across the resistor?

<details>
<summary>Answer</summary>

By KVL: V_resistor = 5V − 0.7V = 4.3V. Then I = V/R = 4.3V / 330Ω = 13.0 mA.

This is exactly the kind of calculation you'll do repeatedly with diodes — the 0.7V component is a forward-biased silicon diode.

*Reference: Module 00, Notebook 01 — Ohm's law and KVL sections*
</details>

---

**Q2.** You have a 9V battery and want exactly 20 mA to flow through a circuit. What value resistor do you need in series (assuming the rest of the circuit drops 2V)?

<details>
<summary>Answer</summary>

The resistor must drop 9V − 2V = 7V. R = V/I = 7V / 0.020A = 350Ω. You'd use a standard 330Ω (giving ~21.2 mA) or 390Ω (giving ~17.9 mA).

This is an LED current-limiting resistor calculation — one of the first things you'll do in Module 02.

*Reference: Module 00, Notebook 01 — Ohm's law section*
</details>

## Kirchhoff's Voltage Law (KVL)

Diode circuits rely heavily on KVL — walking around a loop and accounting for every voltage drop.

---

**Q3.** In a series circuit with a 12V supply, a 470Ω resistor, and a component with an unknown voltage drop, you measure 8.5V across the resistor. What is the voltage across the unknown component?

<details>
<summary>Answer</summary>

KVL: 12V = V_resistor + V_component. So V_component = 12V − 8.5V = 3.5V.

*Reference: Module 00, Notebook 01 — KVL section*
</details>

---

**Q4.** Why is KVL true? State the physical principle behind it in one sentence.

<details>
<summary>Answer</summary>

Energy is conserved: the total energy given to charges by sources must equal the total energy consumed by components around any closed loop.

*Reference: Module 00, Notebook 01 — KVL section*
</details>

## Current Limiting

Many Module 02 experiments involve protecting components with series resistors.

---

**Q5.** An LED has a forward voltage of 2V and a maximum current of 20 mA. You're powering it from 5V. Explain why you need a resistor, what value it should be, and how much power it will dissipate.

<details>
<summary>Answer</summary>

An LED is not a resistor — its forward voltage is roughly constant regardless of current. Without a resistor, current would spike to a destructive level. The resistor must drop the remaining voltage: V_R = 5V − 2V = 3V. R = 3V / 20mA = 150Ω. Power in the resistor: P = I²R = (0.020)² × 150 = 60 mW, or equivalently P = V × I = 3V × 20mA = 60 mW.

*Reference: Module 00, Notebook 01 — LED section*
</details>

## Material Science Foundations

Module 02 dives deep into doping and PN junctions. You need the material science concepts from Module 01.

---

**Q6.** What are **valence electrons**, and why do they determine a material's electrical behavior?

<details>
<summary>Answer</summary>

Valence electrons are the electrons in an atom's outermost shell. They determine electrical behavior because they are the ones available for bonding and conduction. Materials with loosely held valence electrons (1-3) are conductors; materials with tightly held valence electrons (full shell) are insulators; materials with 4 valence electrons (like silicon) are semiconductors.

*Reference: Module 01, Notebook 01 — "Atomic Structure Review" section*
</details>

---

**Q7.** What is the **bandgap**, and what are the approximate bandgap values for a conductor, silicon, and an insulator?

<details>
<summary>Answer</summary>

The bandgap is the energy difference between the valence band (where electrons are bound) and the conduction band (where electrons can move freely). Conductors: ~0 eV (bands overlap). Silicon: 1.12 eV. Insulators: >4 eV (e.g., diamond at 5.5 eV).

*Reference: Module 01, Notebook 01 — "Energy Bands and the Bandgap" section*
</details>

---

**Q8.** A metal wire gets hotter and its resistance increases. A semiconductor chip gets hotter and its resistance decreases. Explain why these go in opposite directions.

<details>
<summary>Answer</summary>

In metals, free electrons are already plentiful. Heat increases lattice vibrations, causing more collisions, so resistance goes **up**. In semiconductors, heat gives more electrons enough energy to jump the bandgap into the conduction band. The increase in carrier count outweighs the increased collisions, so resistance goes **down**.

*Reference: Module 01, Notebook 01 — "Temperature and Conductivity" section*
</details>

---

**Q9.** Silicon has 4 valence electrons. In pure silicon at room temperature, roughly what fraction of silicon atoms have a free electron? (Order of magnitude is fine.)

<details>
<summary>Answer</summary>

About 1 in 10¹³. Silicon has ~5 × 10²² atoms/cm³ and ~1.5 × 10¹⁰ intrinsic carriers/cm³ at room temperature. This is an incredibly small fraction — which is precisely what makes doping so powerful (adding just 1 impurity per million atoms increases conductivity by ~1,000,000×).

*Reference: Module 01, Notebook 01 — carrier concentration plots*
</details>

## Electric Fields

The PN junction creates a built-in electric field that is central to how diodes work.

---

**Q10.** What does an electric field do to a charged particle? If an electron is in an electric field that points to the right, which direction does the electron move?

<details>
<summary>Answer</summary>

An electric field exerts a force on charged particles. Positive charges feel a force in the direction of the field; negative charges feel a force **opposite** to the field. An electron (negative) in a rightward field moves to the **left**.

In a PN junction, the built-in electric field in the depletion region pushes electrons back to the N-side and holes back to the P-side — this is the drift current that opposes diffusion.

*Reference: Module 00, Notebook 01 — "Electric Fields" section*
</details>
