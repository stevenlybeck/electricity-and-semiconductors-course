## Voltage, Current, and Charge

Module 01 discusses how charge carriers move through different materials. You need a solid understanding of what voltage, current, and charge are.

---

**Q1.** What is voltage? Describe it in terms of energy and charge, not just as a number on a multimeter.

<details>
<summary>Answer</summary>

Voltage is the energy per unit charge — the amount of work (in joules) needed to move one coulomb of charge between two points. It is always measured *between* two points; a single point does not have a voltage by itself.

*Reference: Module 00, Notebook 01 — Electrical Fundamentals, "Voltage" section*
</details>

---

**Q2.** Electric current is defined as the flow of charge. But in a metal wire, what actually moves? And does it move in the direction of conventional current, or the opposite direction?

<details>
<summary>Answer</summary>

In a metal wire, **electrons** are the charge carriers that move. They flow from the negative terminal toward the positive terminal, which is the **opposite** direction of conventional current. Conventional current flows from + to −, a historical convention from Benjamin Franklin.

*Reference: Module 00, Notebook 01 — "Current" section*
</details>

---

**Q3.** Are electrons always the charge carriers? Name a situation where something other than electrons carries the current.

<details>
<summary>Answer</summary>

No. In a battery's electrolyte, **ions** (charged atoms or molecules) carry the current. In semiconductors, current is carried by both electrons and **holes** (the absence of an electron that behaves like a positive charge carrier). Module 01 will explore these differences in detail.

*Reference: Module 00, Notebook 01 — "Resistance" section (charge carrier discussion)*
</details>

## Resistance and Ohm's Law

Module 01 introduces **resistivity** — a material property — and relates it to the atomic structure of different materials. You need to be comfortable with resistance as a circuit concept first.

---

**Q4.** State Ohm's law. If a 1 kΩ resistor has 5V across it, how much current flows?

<details>
<summary>Answer</summary>

V = IR, so I = V/R = 5V / 1000Ω = 5 mA.

*Reference: Module 00, Notebook 01 — "Resistance and Ohm's Law" section*
</details>

---

**Q5.** You learned in Module 00 that metals get *more* resistive as they heat up. Can you explain why in one sentence?

<details>
<summary>Answer</summary>

Higher temperature causes the metal lattice to vibrate more, which means free electrons collide more often with vibrating atoms, increasing resistance.

*Reference: Module 00, Notebook 01 — "Resistance" material science section*
</details>

## Electric Fields

Module 01 discusses how electric fields interact with electrons in different materials. Make sure you have a working mental model of what an electric field is.

---

**Q6.** What is an electric field? How is it similar to a gravitational field?

<details>
<summary>Answer</summary>

An electric field is a region of space where a charged particle experiences a force. It is analogous to a gravitational field: just as mass creates a gravitational field that pulls other masses, electric charge creates an electric field that pushes or pulls other charges. The field exists whether or not there is a test charge present to feel it.

*Reference: Module 00, Notebook 01 — "Electric Fields" section*
</details>

---

**Q7.** When you connect a battery to a wire, current begins flowing almost instantly throughout the entire wire, even though individual electrons move very slowly (fractions of a millimeter per second). What propagates quickly, and how fast does it travel?

<details>
<summary>Answer</summary>

The **electric field** (an electromagnetic wave) propagates through and around the wire at roughly 2/3 the speed of light in copper. This field is what pushes all the electrons in the wire simultaneously. The electrons themselves drift slowly, but the signal to "start moving" arrives almost instantly.

*Reference: Module 00, Notebook 01 — "How fast does the field propagate?" section*
</details>

## Power

---

**Q8.** A 220Ω resistor has 9V across it. How much power does it dissipate? Which formula did you use and why?

<details>
<summary>Answer</summary>

P = V²/R = 81/220 ≈ 0.37W (368 mW). This form is most convenient because we know V and R directly. You could also find I first (I = V/R = 40.9 mA) and then use P = IV = 9 × 0.0409 ≈ 0.37W.

*Reference: Module 00, Notebook 01 — "Power" section*
</details>
